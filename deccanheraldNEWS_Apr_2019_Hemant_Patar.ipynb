```python
#import all the required libraries
import pandas as pd, requests, time
import re
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains 
```

```python
# creating empty lists for scraping data
news_v=[]
news_h=[]
news_dt=[]
news_a=[]
news_d=[]  

driver=webdriver.Chrome(executable_path='D:\chromedriver.exe')

for j in range(9):
    driver.get("https://www.deccanherald.com/sitemap/detail/days/2019-04-0{}".format(j+1))
    time.sleep(3)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    heading = soup.find_all("div", {"class": "group parent-relative grid-col-full section-heading-wrapper"})
    news_artcl = soup.find_all('ul', attrs ={'class':'group notab-snippets notab-snippets--graybg f-left full-width notab-snippets--archive-note'})

    for x, i in enumerate(heading):
        li = news_artcl[x].find_all("li", {"class": "group sanspro-reg archives-note"})
        for j in li:
            news_v.append(i.text)
            news_h.append(j.find("a").text) 

            links="https://www.deccanherald.com"+str(j.find("a").get('href'))
            driver.get(links)
            time.sleep(3)
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            
            auth = soup.find("div", {"class": "article-author__time"})
            news_a.append((auth.text).split(",  Apr")[0])       

            auth = soup.find("div", {"class": "article-author__time"})
            news_dt.append(((auth.text).split("updated:")[0]).split("  ")[-1]) 
            
            news = soup.find("div", {"class": "field field-name-body field-type-text-with-summary field-label-hidden"})
            news_d.append(news.text)
            
    
         
for i in range(9,30):
    driver.get("https://www.deccanherald.com/sitemap/detail/days/2019-04-{}".format(i+1))
    time.sleep(3)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    heading = soup.find_all("div", {"class": "group parent-relative grid-col-full section-heading-wrapper"})
    news_artcl = soup.find_all('ul', attrs ={'class':'group notab-snippets notab-snippets--graybg f-left full-width notab-snippets--archive-note'})

    for x, i in enumerate(heading):
        li = news_artcl[x].find_all("li", {"class": "group sanspro-reg archives-note"})
        for j in li:
            news_v.append(i.text)
            news_h.append(j.find("a").text) 

            links="https://www.deccanherald.com"+str(j.find("a").get('href'))
            driver.get(links)
            time.sleep(3)
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            
            auth = soup.find("div", {"class": "article-author__time"})
            news_a.append((auth.text).split(",  Apr")[0])       

            auth = soup.find("div", {"class": "article-author__time"})
            news_dt.append(((auth.text).split("updated:")[0]).split("  ")[-1]) 
            
            news = soup.find("div", {"class": "field field-name-body field-type-text-with-summary field-label-hidden"})
            news_d.append(news.text)


driver.quit()

print(len(news_v))
print(len(news_h))
print(len(news_dt))
print(len(news_a))
print(len(news_d))

print(news_v)
print(news_h)
print(news_dt)
print(news_a)
print(news_d)

 # creating the dataframe from the scraped data 
df=pd.DataFrame({"Vertical":news_v,
                 "Headlines":news_h,
                 "Date":news_dt,
                 "Author":news_a, 
                 "Description":news_d})
df.head()
df.to_csv('deccanheraldnews_apr_2019_HEMANT_PATAR.csv', index_label='index', encoding="utf-8")
```
