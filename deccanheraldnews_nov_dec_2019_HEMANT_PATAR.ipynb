```python
#Scraping Deccanherald Nov & Dec 2019 news archive.
import pandas as pd, requests, time
import re
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options 
import csv
 
row = ['Vertical', 'Headlines', 'Date','Author','Description']
file_object = open('deccanheraldnews_nov_dec_2019_HEMANT_PATAR.csv', 'a', encoding='utf-8', newline="")
writer = csv.DictWriter(file_object, fieldnames=row)
if file_object.tell() == 0:
    writer.writeheader()

chrome_options = Options()
chrome_options.add_argument("--headless")
driver=webdriver.Chrome(executable_path='E:\chromedriver.exe', options=chrome_options)
for k in range(11,13):
    for j in range(31):
        if k==11 and j==30:break
        else:
            if j < 9: driver.get("https://www.deccanherald.com/sitemap/detail/days/2019-{}-0{}".format(k,j+1))
            else: driver.get("https://www.deccanherald.com/sitemap/detail/days/2019-{}-{}".format(k,j+1))
            time.sleep(3)  

            soup = BeautifulSoup(driver.page_source, 'html.parser')

            heading = soup.find_all("div", {"class": "group parent-relative grid-col-full section-heading-wrapper"})
            news_artcl = soup.find_all('ul', attrs ={'class':'group notab-snippets notab-snippets--graybg f-left full-width notab-snippets--archive-note'})

            for x, i in enumerate(heading):
                li = news_artcl[x].find_all("li", {"class": "group sanspro-reg archives-note"})
                for j in li:
                    news_v=(i.text)
                    news_h=(j.find("a").text) 

                    links="https://www.deccanherald.com"+str(j.find("a").get('href'))
                    driver.get(links)
                    print(links)
                    time.sleep(4)
                    soup = BeautifulSoup(driver.page_source, 'html.parser')

                    auth = soup.find("div", {"class": "article-author__time"})
                    try:news_a=((auth.text).split(",  Nov")[0])  
                    except:news_a=((auth.text).split(",  Dec")[0]) 
                    news_dt=(((auth.text).split("updated:")[0]).split("  ")[-1]) 

                    news = soup.find("div", {"class": "field field-name-body field-type-text-with-summary field-label-hidden"})
                    news_d=(news.text)
                    
                    writer.writerow({"Vertical":news_v, "Headlines":news_h, "Date":news_dt, "Author":news_a, "Description":news_d})
file_object.close()
print("CSV creation done!!")

driver.quit()
```
