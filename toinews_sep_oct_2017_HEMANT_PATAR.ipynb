```python
#Scraping toi Sep & Oct 2017 news archive.
import pandas as pd, requests, time
import re
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options 
import csv
```

```python
row = ['date', 'author', 'vertical', 'headline', 'description']
file_object = open('toinews_sep_oct_2017_HEMANT_PATAR.csv', 'a', encoding='utf-8', newline="")
writer = csv.DictWriter(file_object, fieldnames=row)
if file_object.tell() == 0:
    writer.writeheader()

base_counter = [l+1 for l in range(42978,43039)] 
for m in range(9,11):  
    for d in range(32 if m % 2 == 0 else 30):
        r = requests.get("https://timesofindia.indiatimes.com/2017/{}/{}/archivelist/year-2017,month-{},starttime-{}.cms".format(m, d+1, m, base_counter.pop(0)))
        time.sleep(2)  

        soup = BeautifulSoup(r.text, 'html.parser')
        table = soup.find_all("table")[1]
        tr = table.find_all("tr")[2]
        span = tr.find("span")
        a = span.find_all("a")
        details = {}
        for x, j in enumerate(a):
            nl = requests.get(j.get("href"))
            print(j.get("href"))
            time.sleep(2)

            if "Page Not Found" in nl.text:
                print("Page Not Found")
                pass
            elif "this-article-has-been-removed" in str(j.get("href")):
                print("Page Not Found")
                pass
            else:
                try:
                    soup = BeautifulSoup(nl.text, 'html.parser')
                    try:
                        div = soup.find("div", {"class": "as_byline"})
                        details['date'] = str(div.find("div", {"data-plugin": "dateformat"}).text).replace("Updated: ", "").replace("Created: ", "")
                        author = str(div.text).split("-")[1].replace(details['date'], "").replace(" TNN", "").replace("Updated: ", "").replace("Created: ", "").replace("facebooktwitterPintrest", "")
                        details['author'] = author.replace("TOI", "").replace("tnn", "").replace("toi", "").replace("TNn", "").replace("TNN ", "").replace("Tnn", "").replace("TIMESOFINDIA.COM", "").replace(" TIMESOFINDIA.COM", "")
                    except:
                        try:
                            div = soup.find("div", {"class": "_3Mkg- byline"})
                            details['date'] = str(div.text).split("|")[-1].replace("Updated: ", "").replace("Created: ", "").lstrip()
                            if len(str(div.text).split(" | ")) == 2:
                                author = str(div.text).split(" | ")[0].replace("TOI", "").replace("tnn", "").replace("toi", "").replace(" TNN", "").replace("TNn", "").replace("TNN ", "").replace("Tnn", "").replace("TIMESOFINDIA.COM", "").replace(" TIMESOFINDIA.COM", "")
                                details['author'] = "" if author == "TNN" or "IST" in author else author
                            else: details['author'] = "" if "IST" in str(div.text).split(" | ")[0] else str(div.text).split(" | ")[0].replace("TOI", "").replace("tnn", "").replace("toi", "").replace(" TNN", "").replace("TNn", "").replace("TNN ", "").replace("Tnn", "").replace("TIMESOFINDIA.COM", "").replace(" TIMESOFINDIA.COM", "")
                        except:
                            div = soup.find("div", {"class": "byline-content"})
                            details['date'] = str(div.text).split("|")[-1].replace("Updated: ", "").replace("Created: ", "").lstrip()
                            details['author'] = str(div.text).split("|")[0].replace(details['date'], "").replace("TOI", "").replace("tnn", "").replace("toi", "").replace(" TNN", "").replace("TNn", "").replace("TNN ", "").replace("Tnn", "").replace("TIMESOFINDIA.COM", "").replace(" TIMESOFINDIA.COM", "")

                    details['vertical'] = str(j.get("href")).split("/")[4][0].upper() + str(j.get("href")).split("/")[4].replace("-", "")[1:]
                    details['headline'] = j.text

                    try:
                        div = soup.find("div", {"class": "Normal"})
                        details['description'] = str(div.text).replace("\n", "")
                    except:
                        div = soup.find("div", {"class": "ga-headlines"})
                        details['description'] = str(div.text).replace("\n", "")

                    writer.writerow(details)
                except Exception as e: 
                    print(e)
                    pass
file_object.close()
print("CSV creation done!!")
```
